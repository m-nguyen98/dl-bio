```python
import datasets.livecell.livecell as livecell
from backbones.resnet import ResNet34
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
import torch
from torch import nn as nn
import os
_data_dir = 'data/livecell'
```

```python
def load_livecell( mode='train', min_samples=20):
    train_cell_types = ["A172", "BT474", "BV2", "Huh7"]
    val_cell_types = ["MCF7", "SHSY5Y"]
    test_cell_types = ["SkBr3", "SKOV3"]
    split = {'train': train_cell_types,
                'val': val_cell_types,
                'test': test_cell_types}

    cell_types = split[mode]
    
    file_names = []
    labels = []
    for filename in os.listdir(_data_dir):
        file_label = filename.split('_')[0]
        if filename.endswith('tif') and file_label in cell_types:
            file_names.append(filename)
            labels.append(file_label)

    return file_names, labels

samples, targets = load_livecell()
```

```python
def show_images(images, nmax=64):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(make_grid((images[:nmax]), nrow=8).permute(1, 2, 0))

def show_batch(dl, nmax=32):
    for images, labels in dl:
        print(images.size())
        show_images(images, nmax)
        break
```
