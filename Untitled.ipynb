```python
model = ResNet34()

train_dataset = livecell.LCSimpleDataset(4, mode='train')
train_dl = train_dataset.get_data_loader()

val_dataset = livecell.LCSimpleDataset(2, mode='val')
val_dl = val_dataset.get_data_loader()
```

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    # Set the model to training mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practices
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        # Compute prediction and loss
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def train(train_loader, val_loader, model):
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    loss = nn.CrossEntropyLoss()
    max_acc = -1

    for epoch in range(10):
        model.train()
        train_loop(train_loader, model, loss, optimizer)

        if epoch % 2 == 0:
            model.eval()
            acc = model.test_loop(val_loader)
            print(f"Epoch {epoch}: {acc:.2f}")

            if acc > max_acc:
                print("best model! save...")
                max_acc = acc

    return model
```

```python
train(train_dl, val_dl, model)
```
